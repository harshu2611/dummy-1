<b>TaskPY200_T6: Find the Smallest Number With “Digit Constraints”<br>
Input: N <br>
Find smallest number x in 1..N such that: <br>
• sum of digits of x is divisible by 7 <br>
• x contains exactly two 3s <br>
• x does not contain digit 0 <br>
If none, print -1.<br>

```python
# Taking a input number
N = int(input("Enter N: "))

found = 0   # Flag to check if answer exists

for i in range(1, N + 1): # N+1 because loop runs till (stop-1)

    digit = i
    digit_sum = 0
    count_3 = 0
    contains_zero = 0

    # Extracting digits
    while digit > 0:
        sub_digit = digit % 10

        # Check zero digit
        if sub_digit == 0:
            contains_zero = 1

        # Count number of 3s
        if sub_digit == 3:
            count_3 = count_3 + 1

        # Add digit to sum
        digit_sum = digit_sum + sub_digit

        digit = digit // 10  # floor the value of temp

    # Check all conditions
    if digit_sum % 7 == 0 and count_3 == 2 and contains_zero == 0:
        print(i)
        found = 1
        break

# If no such number found
if found == 0:
    print(-1)
```
