<b>TaskPY200_T5: Chessboard with Numbers <br>
Input: N <br>
Print an N x N grid:<br> 
• If (row+col) is even → print 1 <br>
• else → print 0<br> 
But if row == col → print X instead.<br> 

```python
# Take a number of grids from user
N = int(input("Enter N:"))

for i in range(1,N+1):              # i means row number
    for j in range(1,N+1):          # j means column number

        # If (i == j) --> print X
        if i==j:
            print("X" , end=" ")

        # If (i + j) is even --> print 1   
        elif (i+j)%2==0:
            print("1" , end=" ")

        # Otherwise --> print 0    
        else:
            print("0" , end=" ")
    # to change the line of each row
    print()
```
