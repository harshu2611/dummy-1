TaskPY200_T4: Write program to check password strength rules <br>
Input a string s (assume lowercase/uppercase/digits may appear) <br>
Now check: if the string contains full fills and print final output: valid (STRONG) or Invalid
password (WEAK)<br>
1. length >= 8 <br>
2. has at least 1 digit <br>
3. has at least 1 uppercase <br>
4. has at least 1 lowercase <br>
Print “STRONG” else “WEAK”<br>

```python
# Taking input of password from user
s = input("Enter your password : ")

# initialize the variables
total_length = 0
contain_digit = 0
contain_uppercase = 0
contain_lowercase = 0


# Calculate total length manually also check characters

for character in s:
    total_length = total_length + 1

    # Check for digit (0–9)
    if character >= '0' and character <= '9':
        contain_digit = 1

    # Check for uppercase (A–Z)
    elif character >= 'A' and character <= 'Z':
        contain_uppercase = 1

    # Check for lowercase (a–z)
    elif character >= 'a' and character <= 'z':
        contain_lowercase = 1

# Check all conditions

if total_length >= 8 and contain_digit == 1 and contain_uppercase == 1 and contain_lowercase == 1:
    print("STRONG")
else:
    print("WEAK")
```
