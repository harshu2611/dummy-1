```python
# Taking input from user
n = int(input("Enter a number : "))

# Oppostive of entered number
neg_n = -(n)

# check positive value
if n > 0:
    print(f"The number you entered {n} is positive, however negative value of your number is {neg_n} ")

# check zero
elif n ==0:
    print("The number you entered is Zero")

# check negative value
else:
    print(f"The number you entered {n} is negative, however positive value of your number is {neg_n} ")
```
