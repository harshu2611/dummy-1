<b>TaskPY200_T3: Write program to calcite Electricity Bill Slabs. User enters a bill unit and 
gets the total bill according to following logic. <br>
Input: units <br>
Rates: <br>
• First 100 @ 2/unit <br>
• Next 100 (101–200) @ 3/unit <br>
• Next 300 (201–500) @ 5/unit <br>
• Above 500 @ 8/unit <br>
Add fixed charge 50 if units > 0 <br>
Print total bill <br>

```python
units = int(input("Enter total units used : "))


total_bill = 0


if units > 0:

    # For First 100 units
    if units <= 100:
        total_bill = units * 2

    # For next 101–200 units
    elif units >100 and units <=200:
        total_bill = (100 * 2) + ((units - 100) * 3)

    # For next 201–500 units
    elif units >200 and units <= 500:
        total_bill = (100 * 2) + (100 * 3) + ((units - 200) * 5)

    # For all Above 500 units
    else:
        total_bill = (100 * 2) + (100 * 3) + (300 * 5) + ((units - 500) * 8)

    # Add fixed charge to total bill
    total_bill = total_bill + 50

print(f"Total Bill to pay is {total_bill}")
```
